```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np        
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
ind=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
ind
```

```python
dep=dataset[['charges']]
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ind,dep,test_size=0.30,random_state=0)
```

```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
preinput=sc.transform(([[1234,345,4565,1,0,]]))
```

```python
preinput
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={"kernel":["rbf","sigmoid","linear","poly"],"C":[10,200,300,1000,3000],"gamma":["auto","scale"]}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
re=grid.cv_results_
print("the best parameters of tuning{}:".format(grid.best_params_))
```

```python

table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=1000)
regressor.fit(x_train,y_train)
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```

```python
regressor.intercept_
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
